```python
def palindrome_string_finder(string:str=None)->bool:
    if string==None:
        raise TypeError("Function can't be empty")
    if not isinstance(string,str):
        raise TypeError("Input must be in String format")
    reverse=""
    for item in string:
        if item==" ":
            raise TypeError("Input can't contain whitespaces")
        reverse= item+reverse
    return reverse  == string
    
if __name__== "__main__":
    try:
        palindrome="madam"
        result = palindrome_string_finder(palindrome)
        if result== True:
            print(f"{palindrome} is palindrome String")
        else:
             print(f"{palindrome} is not a palindrome String")
    except (ValueError ,TypeError) as err:
        print("Input must be a String and can't be empty")
    except Exception as e:
        print(e)
```

```python
print(palindrome_string_finder("mam"))
```
