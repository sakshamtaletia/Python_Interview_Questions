```python
def list_of_prime_number(staring_num :int=None , ending_num:int=None)->list:
    if staring_num==None or ending_num==None:
        raise TypeError("Function can't be empty")
    if not isinstance(staring_num or ending_num,int):
        raise TypeError("Input must be an Integer")
    if staring_num<2 :
        raise ValueError("your Starting number must greater then 2")
    if ending_num<staring_num:
        raise ValueError("your Ending number must be greater then Starting number")
    list1=[]
    for item in range(staring_num,ending_num+1):
        for i in range(2,int(item**.5)+1):
            if item%i==0:
                break
        else:
            list1.append(item)
    return list1 
if __name__=="__main__":
    try:
        num1=input("Enter your staring number")
        num2=input("Enter your Ending number")
        if not num1.strip().isdigit():
            raise TypeError("Input must be a valid interger and It can't be empty")
        if not num2.strip().isdigit():
            raise TypeError("Input must be a valid interger and It can't be empty")
        result=list_of_prime_number(int(num1),int(num2))
        print(result)
    except (ValueError , TypeError) as err:
        print(err)
    except Exception as e:
        print(e)
```
