```python
def largest_number(obj:list)-> int:
    """
    Finding the largest number a list contain
    Argument:
            takes a integer type list
    Return : 
            integer value

    """
    if not isinstance(obj,list):
        raise TypeError("your input must be a list of integer ")
    if len(obj)==0:
        raise ValueError("your list is empty")
    temp=0
    for item in obj:
        if isinstance(item,int):
            if item > temp:
                temp=item
        else:
            raise TypeError("your list contain only integer")
    return temp

if __name__== "__main__":
    try:
        # user_input=[123,23,43,32,34,234,213,234,23,-1]
        # result = largest_number(user_input)
        
        # or
        user_input=input("Enter numbers in the list seperate by space")
        number_list=[int(item) for item in user_input.split()]
        result=largest_number(number_list)
        
        print("largest number is : ", result)
    except TypeError as txtErr:
        print(txtErr)
    except Exception as e:
        print(e)

        
     
    
```
