```python
def count_vowel(string : str=None)->int:
    """
    Count the number of vowels present in a string

    Arguments:
             Takes String only
    Return:
          Integer value
          
    """
    if not isinstance(string,str):
        raise TypeError("Input must be a string value")
    vowels="aeiou"
    count=0
    for item in string.lower():
        if item in vowels:
            count+=1
            
    return count

if __name__=="__main__":
    try:
        string = input("Enter the string")
        result = count_vowel(string)
        print(f"Total number of vowels in a string are : {result}")
    except TypeError as err:
        print(err)
    except Exception as e :
        print(e)

    
```
