```python
def reverse_string(string :str)-> str:
    """
    Reversing a string without using any builtin method

    Arguments : 
              Takes only String Value
    
    Return: 
           string Value
    """
    if not isinstance(string,str):
        raise TypeError("Your Input Must Be String")
    if len(string)==0:
        raise ValueError("your data is empty")
    rev=""
    for char in string:
        rev=char+rev
    return rev

if __name__ =="__main__":
    try:
        user_input=input("Enter the Name you want to reverse")
        result=reverse_string(user_input)
        print(result)
    except ValueError as txtError:
        print(txtError)
    except Exception as e:
        print(e)
```
