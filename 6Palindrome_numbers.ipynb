```python
""" Find the palindrome number like 121 , 12321 etc"""
```

```python
num = int(input("Enter the number "))
palindrome=num
temp=0
while num>0:
    r=  num%10
    # print("r",r)
    temp= temp*10 +r
    # print("temp",temp)
    num=int(num/10)
    # print("num",num)
if temp==palindrome:
    print(f"{palindrome} is palindrome number")
else:
    print(f"{palindrome} is not palindrome number")
    
```

```python
def palindrome_finder(number : int=None)->bool:
    """
    Arguments:
             Must be an integer value
    Return : 
            Boolean values
    Function Work:
                 To find whether a number is palindrome or not
                 
    """
    if number==None:
        raise TypeError("Function can't be empty")
    if not isinstance(number ,int):
        raise TypeError("Input must be an integer value")
    num = number
    palindrome=num
    temp=0
    while num>0:
        r=  num%10
        temp= temp*10 +r
        num=int(num/10)
    if temp==palindrome:
        return True
    else:
        return False

if __name__== "__main__":
    try:
        num= input("Enter your number : ")
        if not num.strip().isdigit():
            raise TypeError("Input must be an positive Integer value")
        result = palindrome_finder(int(num))
        if result==True:
            print(f"{num} is palindrome number")
        else:
            print(f"{num} is not palindrome number")
    except (ValueError , TypeError) as err:
        print("Input can't be a non-positive value")
    except Exception as e:
        print(e)

    
```
