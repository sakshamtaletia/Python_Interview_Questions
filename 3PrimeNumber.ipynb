```python
def prime_number(num :int)->bool:
    
    if not isinstance(num,int):
        raise TypeError("num should be an integer")
    if num==2:
        return f"{num} is a prime number"
    elif num<2:
        raise ValueError("your nummber should be 2 or greater then it")
    for item in range(2,int(num**.5)+1):
        if num%item==0:
            return False
            break
    else:
        return True
            
if __name__  == "__main__":
    try:
        # user_input=10007
        # result=prime_number(user_input)
        # print(result)

        # or

        user_input=input("Enter the number" )
        if not user_input.strip().isdigit():
            raise TypeError(" input must be a digit ")
        number=int(user_input)
        if prime_number(number):
            print(f"{number} is a prime number")
        else:
            print(f"{number} is not a prime number")
            
    except TypeError as txtErr:
        print(txtErr)
    except Exception as e:
        print(e)
            
```
