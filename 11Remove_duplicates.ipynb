```python
""" Remove Duplicate element from a list """
```

```python
list1=[1,1,2,34,35,35,3,4,5,6,6,7,"s",'s' , 't','t','1','1']
final_list=[]
for item in list1:
    if item not in final_list:
        final_list.append(item)
print(final_list)
```

```python
def remove_duplicate(list1 : list=None)->list:
    """
    Remove duplicate elements from a list
    Arguments:
            A list can contain any type of element
    Return:
          list with unique elements
          
    """
    if list1==None:
        raise ValueError("list can't be empty")
    if not isinstance(list1,list):
        raise TypeError("Input must be a list")
    if len(list1)==1:
        print("list must contain more then 1 element so function can remove duplicates")
        return list1
        
    final_list=[]
    for item in list1:
        if item not in final_list:
            final_list.append(item)
    return final_list


if __name__=="__main__":
    try:
        user_input=[1,1,2,34,35,35,3,4,5,6,6,7,"saksham",'saksham' , 't','t','1','1']
        result = remove_duplicate(user_input)
        print(result)

    except (ValueError , TypeError) as err:
        print(err)
    except Exception as e:
        print("Error",e)
```
