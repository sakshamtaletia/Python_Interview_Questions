```python
""" find factorial of a function using recursion """
```

```python

def factorial_finder(num:int=None)->int:
    """
    Calculate the factorial

    Arguments:
             Take an Integer value
    Return:
           Only interger after calculating the factorial

           
    """
    if num==None :
        raise TypeError("Number can't be empty ")
    if not isinstance(num , int) or num<0:
        raise ValueError("Input must be a not-negative Integer value")
    if num==0 or num ==1:
        return 1
    return num*fact(num-1)

if __name__=="__main__":
    try:
        user_input=input("Enter the number : ")
        if not user_input.strip().isdigit():
            raise ValueError("Input must be a not-negative Integer value")
        result=factorial_finder(int(user_input))
        print(result)
    except (ValueError , TypeError)  as err:
        print("Input Can't be empty and must be non-negative integer")
    except Exception as e:
        print(e)
    
```
