```python
def second_largest(list1:list)->int:
    """
    Find the second largest number amoung a list

    Argument:
            Take list of integers
    Return:
           Second largest number type integer
    """
    if not isinstance(list1,list):
        raise TypeError("Input must be a list")
        
    for item in list1:
        if not type(item)==int :
            raise TypeError("list must contain only number")
        
    for i in range(0,len(list1)):
        for j in range(0, len(list1)-i-1):
            if list1[j] > list1[j+1]:
                # swap
                temp = list1[j]
                list1[j] = list1[j+1]
                list1[j+1] = temp
    return list1[len(list1)-2]

if __name__=="__main__":
    try:
        user_input=input("Enter the number seprate the by white space")
        number_list= [int(item) for item in user_input.split()]
        result =second_largest(number_list)
        print(result)

    except TypeError as err:
        print("Input must be a numbers")
    except Exception as e:
        print(e)
```
