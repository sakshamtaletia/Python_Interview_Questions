```python
def sum_of_digit(number :int=None)->int:
    """
     Find the Sum of digit of a number
    Arguments:
              Take Integer only
    Return : 
           Integer: Sum of digits
           
    """
    if number==None:
        raise TypeError("Function contain Some Integer value")
    if not isinstance(number ,int) or number<0:
        raise TypeError("Input must be an non-negative integer value")
    sum=0
    while number>0:
        r = number%10
        sum = sum+r
        number=int(number/10)
    return sum

if __name__=="__main__":
    try:
        user_Input = input("Enter the number")
        if not user_Input.strip().isdigit():
            raise ValueError("Input must be an non-negative integer value")
        result = sum_of_digit(int(user_Input))
        print(result)
    except (TypeError ,ValueError) as err:
        print(err)
    except Exception as e:
        print(e)
    
    
```
