```python
"""  Find factorial of a number """
```

```python
num = int(input("Enter your number"))
fact=1
for item in range(1,num+1):
    fact*=item
print(fact)
```

```python
def factorial(num:int=None)->int:
    """
    Calculate the factorial of a non-negative Integer
    
    Argument:
             Num(int): the number for which factorial required

    Return:
           int: factorial of given number
    
    """
    if num is None:
        raise TypeError("Your Function contain atleast one argument")
    if num<0:
        raise ValueError("Input must be a non-negative integer")
    if num==0 or num==1:
        return 1
   
    fact=1
    for item in range(1,num+1):
        fact*=item

    return fact
if __name__=="__main__":
    try:
        number=input("Enter your number ")
        # strip remove the all whitespace character and isdigit check whether all character in string are digits
        if not number.strip().isdigit():
            raise TypeError("Your Input must be a Integer")
        # typecasting string into integer 
        result=factorial(int(number))
        print(result)
    except (TypeError, ValueError)as err:
        print(err)
    except Exception as e:
        print(e)
        
```
